```python
def find_exons_with_codons(gff_file, gene_name):
    exons = []
    with open(gff_file, 'r') as gff:
        for line in gff:
            if line.startswith('#'):
                continue  # Skip comments in the GFF file
            fields = line.strip().split('\t')
            if len(fields) < 9:
                continue  # Skip lines that don't have enough fields
            feature_type = fields[2]
            attributes = fields[8]
            
            if feature_type == 'exon':
                attributes_dict = dict(item.strip().split('=') for item in attributes.split(';'))
                exon_gene_name = attributes_dict.get('gene')
                
                if exon_gene_name == gene_name:
                    exons.append({
                        'chromosome': fields[0],
                        'start': int(fields[3]),
                        'end': int(fields[4])
                    })

    return exons

gene_name = 'DICER1'
gff_file = 'C:/Users/yjara/Documents/Git folders/Storage/ncbi_dataset/data/GCF_000001405.40/genomic.gff'

exons_with_codons = find_exons_with_codons(gff_file, gene_name)

for exon in exons_with_codons:
    print(f"Chromosome: {exon['chromosome']}, Start: {exon['start']}, End: {exon['end']}")
```
